## Importing the dataset into a dataframe
Importing _pandas_ and _seaborn_ packages

```python
import pandas as pd
import seaborn as sns
```

Load an xls file into a dataframe(object), df in short

```python
df = pd.read_excel("https://www.dosm.gov.my/v1/uploads/files/4_Portal%20Content/2_%20Statistics/Open%20Data/Environment/Forested%20and%20Non-Forested%20Areas%2C%20Malaysia%2C%202000%20-%202017.xlsx")
```

## Plotting the dataframe
Show only _Forested_ in _Category_ and on _2017_ then order by _Hectares_ from largest to smallest

```python
plot = df[(df["Category"] == "Forested") & (df["Year"] == 2017)].sort_values(by = "Hectares", ascending = False)
plot
```

```python
ax = sns.barplot(data = plot, x= "State", y = "Hectares" )
```

We can see Sarawak having the largest Forested land and KL having the smallest. So be grateful to your carbon sinks!
